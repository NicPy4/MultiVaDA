    <h1 style="color: #004080; font-size: 36px;">Kalman Filters</h1>
  </div>

  <!-- Notation Section with Two Columns and Background Block -->
  <div class="fragment fade-in" style="background-color: #e3e7ea; padding: 10px 15px; border-radius: 10px; box-shadow: 0px 4px 8px rgba(0,0,0,0.1); margin-bottom: 10px; ">
  <p style="margin: 0;"><strong style="color: #0073e6; font-size: 18px;">Notations</strong></p>
  <div style="display: flex; justify-content: space-between; margin-top: 5px;">

  <!-- First Column Wrapper -->
  <div style="width: 45%; font-size: 18px; color: #424242; line-height: 1.4;">
      <p style="margin: 5px 0;"><strong> $\mathbf{x}_k$:</strong> State estimate at step $k$</p>
      <p style="margin: 5px 0;"><strong> $\mathbf{P}_k$:</strong> State covariance matrix at step $k$</p>
      <p style="margin: 5px 0;"><strong> $\mathbf{F}$:</strong> State transition matrix</p>
      <p style="margin: 5px 0;"><strong> $\mathbf{Q}$:</strong> Process noise covariance matrix</p>
    </div>

  <!-- Second Column Wrapper -->
  <div style="width: 45%; font-size: 18px; color: #424242; line-height: 1.4;">
      <p style="margin: 5px 0;"><strong> $\mathbf{z}_k$:</strong> Measurement at step $k$</p>
      <p style="margin: 5px 0;"><strong> $\mathbf{H}$:</strong> Observation matrix</p>
      <p style="margin: 5px 0;"><strong> $\mathbf{R}$:</strong> Measurement noise covariance matrix</p>
      <p style="margin: 5px 0;"><strong> $\mathbf{K}_k$:</strong> Kalman gain at step $k$</p>
    </div>

  </div>
</div>   

  <!-- Predict Step -->
<div class="fragment fade-in" style="background-color: #f5f5f5; padding: 10px 15px; border-radius: 10px; box-shadow: 0px 4px 8px rgba(0,0,0,0.1); font-size: 18px; line-height: 1.2; margin-bottom: 10px;">
  <p style="margin: 0;"><strong style="color: #0073e6;">Predict Step</strong></p>
  <p style="margin: 5px 0;">State prediction: $\mathbf{x}_{k|k-1} = \mathbf{F} \, \mathbf{x}_{k-1|k-1} + \mathbf{B} \, \mathbf{u}_k$</p>
  <p style="margin: 5px 0;">Covariance prediction: $\mathbf{P}_{k|k-1} = \mathbf{F} \, \mathbf{P}_{k-1|k-1} \, \mathbf{F}^T + \mathbf{Q}$</p>
</div>

<!-- Update Step -->
<div class="fragment fade-in" style="background-color: #e3f2fd; padding: 10px 15px; border-radius: 10px; box-shadow: 0px 4px 8px rgba(0,0,0,0.1); font-size: 18px; line-height: 1.2;">
  <p style="margin: 0;"><strong style="color: #0073e6;">Update Step</strong></p>
  <p style="margin: 5px 0;">Kalman Gain: $\mathbf{K}_k = \mathbf{P}_{k|k-1} \, \mathbf{H}^T \left( \mathbf{H} \, \mathbf{P}_{k|k-1} \, \mathbf{H}^T + \mathbf{R} \right)^{-1}$</p>
  <p style="margin: 5px 0;">State Update: $\mathbf{x}_{k|k} = \mathbf{x}_{k|k-1} + \mathbf{K}_k \left( \mathbf{z}_k - \mathbf{H} \, \mathbf{x}_{k|k-1} \right)$</p>
  <p style="margin: 5px 0;">Covariance Update: $\mathbf{P}_{k|k} = \left( \mathbf{I} - \mathbf{K}_k \, \mathbf{H} \right) \, \mathbf{P}_{k|k-1}$</p>
</div>


</div>


<div style="background-color: #f0f4f8; padding: 30px; border-radius: 10px; box-shadow: 0px 4px 10px rgba(0,0,0,0.1); margin: 0 100px;"> 
    
  <div style="text-align: center; padding-bottom: 15px;">
    <h1 style="color: #004080; font-size: 36px;">Process Noise Covariance (Q)</h1>
  </div>

  <!-- Introductory Question -->
  <p class="fragment fade-in" style="text-align: center; color: #0073e6; font-size: 22px; font-weight: bold; margin-top: 0; margin-bottom: 20px;">
    How much do we trust our model?
  </p>

  <!-- Explanatory Section with Background -->
  <div class="fragment fade-in" style="font-size: 20px; background-color: #e3f2fd; color: #42242;; line-height: 1.6; border-radius: 10px; box-shadow: 0px 4px 8px rgba(0,0,0,0.1); padding: 20px; ">
    <p style="margin-top: 0;">The world we live in is not perfect... Same with our model predictions!</p>
    <p> The <strong>Q matrix</strong> accounts for how much <strong>uncertainty</strong> there is in the <strong>process model</strong> itself. In other words, Q describes how much unexpected "wiggle room" there is in the prediction due to unmodeled parameters or random influences.</p>
  </div>

  <!-- Key Points Section -->
  <div style="margin-top: 20px; font-size: 20px; color: #424242; line-height: 1.6;">
    <ul style="padding-left: 0; list-style-type: none;">
      <li class="fragment fade-in"  style="margin-bottom: 15px;">
        <strong style="color: #d9534f;">üî∫ If Q is large:</strong> This makes the filter more skeptical of its predictions and more reliant on measurements to correct itself.
      </li>
      <li class="fragment fade-in" >
        <strong style="color: #5cb85c;">üîπ If Q is small:</strong> We‚Äôre fairly confident in our process model. The filter will trust its own predictions more and won‚Äôt adjust the state much based on measurements.
      </li>
    </ul>
  </div>

</div>


<div style="background-color: #f0f4f8; padding: 30px; border-radius: 10px; box-shadow: 0px 4px 10px rgba(0,0,0,0.1); margin: 0 100px;"> 
    
  <div style="text-align: center; font-family: Arial, sans-serif; padding-bottom: 15px;">
    <h1 style="color: #004080; font-size: 36px;">Measurement Noise Covariance (R)</h1>
  </div>

  <!-- Introductory Question -->
  <p class="fragment fade-in" style="text-align: center; color: #0073e6; font-size: 22px; font-weight: bold; margin-top: 0; margin-bottom: 20px;">
    How much do we trust our measurements?
  </p>

  <!-- Explanatory Section with Background -->
  <div class="fragment fade-in" style="font-size: 20px; background-color: #e3f2fd; color: #42242;; line-height: 1.6; border-radius: 10px; box-shadow: 0px 4px 8px rgba(0,0,0,0.1); padding: 10px; ">
    <p style="line-height:1.2em"> 
      The <strong>R matrix</strong> represents measurement noise. It tells the filter how much to trust the measurements</p>
  </div>  
    
  <!-- Key Points Section -->
  <div style="margin-top: 20px; font-size: 20px; color: #424242; line-height: 1.6;">
    <ul style="padding-left: 0; list-style-type: none;">
      <li class="fragment fade-in" style="margin-bottom: 15px;">
        <strong style="color: #d9534f;">üî∫ If R is large:</strong> Indicates high uncertainty in measurements. The filter becomes less trusting of new measurements and relies more on the process model.
      </li>
      <li class="fragment fade-in" >
        <strong style="color: #5cb85c;">üîπ If R is small:</strong> Means measurements are relatively accurate and reliable. The filter will prioritize the measurements and rely less on its prediction.
      </li>
    </ul>
  </div>

</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
import ipywidgets as widgets

# Generate synthetic data for position with noise
np.random.seed(42)
num_steps = 50
true_position = np.linspace(0, 1, num_steps)
measurement_noise = np.random.normal(0, 0.1, num_steps)
measurements = true_position + measurement_noise

# Kalman filter function with 2-sigma confidence interval
def kalman_filter_live_with_uncertainty(Q, R):
    x_est = np.mean(measurements)  # Initial position estimate based on mean
    P = 0.5  # Adjusted initial estimate covariance
    estimates = []
    uncertainties = []  # To store uncertainties (2 standard deviations)

    for z in measurements:
        # Prediction step
        P += Q

        # Update step
        K = P / (P + R)  # Kalman gain
        x_est = x_est + K * (z - x_est)
        P = (1 - K) * P
        uncertainty = 2 * np.sqrt(P)  # 2-sigma for 95% CI

        estimates.append(x_est)
        uncertainties.append(uncertainty)

    # Plot with 2-sigma confidence intervals
    estimates = np.array(estimates)
    uncertainties = np.array(uncertainties)

    plt.figure(figsize=(9, 6))
    plt.plot(true_position, 'k-', label='True Position')
    plt.plot(measurements, 'k.', label='Measurements', alpha=0.5)
    plt.plot(estimates, label=f'Estimates (Q={Q}, R={R})')
    plt.fill_between(range(num_steps),
                     estimates - uncertainties,
                     estimates + uncertainties,
                     color='blue', alpha=0.2, label='Uncertainty (¬±2œÉ)')
    plt.xlabel('Time Step')
    plt.ylabel('Position')
    plt.title('Effect of Q and R on Kalman Filter Estimates with 2-Sigma Uncertainty')
    plt.legend()
    plt.show()
```

```python
interact(kalman_filter_live_with_uncertainty,
         Q=widgets.FloatLogSlider(value=1e-4, base=10, min=-6, max=-2, step=0.5, description='Q'),
         R=widgets.FloatLogSlider(value=1e-3, base=10, min=-4, max=-1, step=0.5, description='R'),
         __display__=False)
```

```python
# Simulation parameters
dt = 0.1  # time step
num_steps = 400  # number of time steps
measurement_noise_std = 2.0  # standard deviation of measurement noise
initial_process_noise_std_position = 0.1  # initial process noise std for manual tuning
initial_process_noise_std_velocity = 0.1  # initial process noise std for manual tuning

# Ground truth velocity model
def true_velocity(t):
    return 10 * np.sin(0.1 * t)  # Example sine-based velocity function

# Generate true position, velocity, and noisy measurements
def generate_data(num_steps, dt, measurement_noise_std):
    true_positions, true_velocities, measurements = [], [], []
    x_true, v_true = 0.0, true_velocity(0)
    for t in range(num_steps):
        true_positions.append(x_true)
        true_velocities.append(v_true)
        measurements.append(x_true + np.random.normal(0, measurement_noise_std))
        x_true += v_true * dt
        v_true = true_velocity(t * dt)
    return np.array(true_positions), np.array(true_velocities), np.array(measurements)
```

```python
# Initial matrices and parameters
F = np.array([[1, dt], [0, 1]])  # State transition matrix
H = np.array([[1, 0]])           # Measurement matrix

# Kalman filter function
def kalman_filter(Q, R, measurements):
    x_hat, P = np.array([0, 0]), np.eye(2)  # Initial state and covariance
    estimated_positions, estimated_velocities = [], []
    residuals = []
    for z in measurements:
        # Prediction step
        x_hat = F @ x_hat
        P = F @ P @ F.T + Q

        # Measurement update step
        y = z - (H @ x_hat)
        S = H @ P @ H.T + R
        K = P @ H.T @ np.linalg.inv(S)
        x_hat = x_hat + K.flatten() * y  # Ensure K is 1D
        P = (np.eye(2) - K @ H) @ P

        estimated_positions.append(x_hat[0])
        estimated_velocities.append(x_hat[1])
        residuals.append(y)

    return np.array(estimated_positions), np.array(estimated_velocities), np.array(residuals)

# RMSE calculation function
def calculate_rmse(estimated, true):
    return np.sqrt(np.mean((estimated - true) ** 2))
```

```python
# Method 1: Manual Tuning (Fixed)
Q_manual = np.array([[initial_process_noise_std_position**2, 0],
                     [0, initial_process_noise_std_velocity**2]])
R_manual = np.array([[measurement_noise_std**2]])

# Method 2: Empirical Estimation (Fixed)
def empirical_estimation_qr_batch(measurements, initial_Q, initial_R):
    est_positions, est_velocities, residuals = kalman_filter(initial_Q, initial_R, measurements)
    R_empirical = np.var(residuals)
    Q_empirical = initial_Q  # Keeping Q fixed for simplicity
    return Q_empirical, np.array([[R_empirical]])

# Method 3: Adaptive Filtering (Dynamic)
def kalman_filter_adaptive(measurements, initial_Q, initial_R, alpha=0.1):
    x_hat, P = np.array([0, 0]), np.eye(2)  # Initial state and covariance
    estimated_positions, estimated_velocities = [], []
    residuals = []
    R = initial_R.copy()
    
    for z in measurements:
        # Prediction step
        x_hat = F @ x_hat
        P = F @ P @ F.T + initial_Q

        # Measurement update step
        y = z - (H @ x_hat)
        S = H @ P @ H.T + R
        K = P @ H.T @ np.linalg.inv(S)
        x_hat = x_hat + K.flatten() * y  # Ensure K is 1D
        P = (np.eye(2) - K @ H) @ P

        estimated_positions.append(x_hat[0])
        estimated_velocities.append(x_hat[1])
        residuals.append(y)

        # Adaptive Update of R
        R = (1 - alpha) * R + alpha * np.array([[y**2]])

    return np.array(estimated_positions), np.array(estimated_velocities), np.array(residuals)

# Method 4: Expectation-Maximization (Iterative)
def expectation_maximization_qr(em_iterations, measurements, initial_Q, initial_R):
    Q_em, R_em = initial_Q.copy(), initial_R.copy()
    
    for _ in range(em_iterations):
        # E-step: Run Kalman Filter with current Q and R
        est_positions, est_velocities, residuals = kalman_filter(Q_em, R_em, measurements)
        
        # M-step: Update R based on residuals
        R_em = np.var(residuals)
        
        # Optionally, update Q here based on state estimates
        # For simplicity, we keep Q fixed
    return Q_em, np.array([[R_em]])
```

```python
true_positions, true_velocities, measurements = generate_data(num_steps, dt, measurement_noise_std)

# Run Kalman filters with each method
results = {}

# 1. Manual Tuning
est_positions_manual, est_velocities_manual, _ = kalman_filter(Q_manual, R_manual, measurements)
position_rmse_manual = calculate_rmse(est_positions_manual, true_positions)
velocity_rmse_manual = calculate_rmse(est_velocities_manual, true_velocities)
results["Manual Tuning"] = {
    "estimated_positions": est_positions_manual,
    "estimated_velocities": est_velocities_manual,
    "position_rmse": position_rmse_manual,
    "velocity_rmse": velocity_rmse_manual
}
print(f"Manual Tuning - Position RMSE: {position_rmse_manual:.3f}, Velocity RMSE: {velocity_rmse_manual:.3f}")

# 2. Empirical Estimation
Q_empirical, R_empirical = empirical_estimation_qr_batch(measurements, Q_manual, R_manual)
est_positions_empirical, est_velocities_empirical, _ = kalman_filter(Q_empirical, R_empirical, measurements)
position_rmse_empirical = calculate_rmse(est_positions_empirical, true_positions)
velocity_rmse_empirical = calculate_rmse(est_velocities_empirical, true_velocities)
results["Empirical Estimation"] = {
    "estimated_positions": est_positions_empirical,
    "estimated_velocities": est_velocities_empirical,
    "position_rmse": position_rmse_empirical,
    "velocity_rmse": velocity_rmse_empirical
}
print(f"Empirical Estimation - Position RMSE: {position_rmse_empirical:.3f}, Velocity RMSE: {velocity_rmse_empirical:.3f}")

# 3. Adaptive Filtering
est_positions_adaptive, est_velocities_adaptive, _ = kalman_filter_adaptive(measurements, Q_manual, R_manual)
position_rmse_adaptive = calculate_rmse(est_positions_adaptive, true_positions)
velocity_rmse_adaptive = calculate_rmse(est_velocities_adaptive, true_velocities)
results["Adaptive Filtering"] = {
    "estimated_positions": est_positions_adaptive,
    "estimated_velocities": est_velocities_adaptive,
    "position_rmse": position_rmse_adaptive,
    "velocity_rmse": velocity_rmse_adaptive
}
print(f"Adaptive Filtering - Position RMSE: {position_rmse_adaptive:.3f}, Velocity RMSE: {velocity_rmse_adaptive:.3f}")

# 4. EM Algorithm
em_iterations = 5  # Number of EM iterations
Q_em, R_em = expectation_maximization_qr(em_iterations, measurements, Q_manual, R_manual)
est_positions_em, est_velocities_em, _ = kalman_filter(Q_em, R_em, measurements)
position_rmse_em = calculate_rmse(est_positions_em, true_positions)
velocity_rmse_em = calculate_rmse(est_velocities_em, true_velocities)
results["EM Algorithm"] = {
    "estimated_positions": est_positions_em,
    "estimated_velocities": est_velocities_em,
    "position_rmse": position_rmse_em,
    "velocity_rmse": velocity_rmse_em
}
print(f"EM Algorithm - Position RMSE: {position_rmse_em:.3f}, Velocity RMSE: {velocity_rmse_em:.3f}")

# Visualization of results
time = np.arange(0, num_steps * dt, dt)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation parameters
dt = 0.1  # time step
num_steps = 400  # number of time steps
measurement_noise_std = 2.0  # standard deviation of measurement noise
initial_process_noise_std_position = 0.1  # initial process noise std for manual tuning
initial_process_noise_std_velocity = 0.1  # initial process noise std for manual tuning

# Ground truth velocity model
def true_velocity(t):
    return 10 * np.sin(0.1 * t)  # Example sine-based velocity function

# Generate true position, velocity, and noisy measurements
def generate_data(num_steps, dt, measurement_noise_std):
    true_positions, true_velocities, measurements = [], [], []
    x_true, v_true = 0.0, true_velocity(0)
    for t in range(num_steps):
        true_positions.append(x_true)
        true_velocities.append(v_true)
        measurements.append(x_true + np.random.normal(0, measurement_noise_std))
        x_true += v_true * dt
        v_true = true_velocity(t * dt)
    return np.array(true_positions), np.array(true_velocities), np.array(measurements)

true_positions, true_velocities, measurements = generate_data(num_steps, dt, measurement_noise_std)

# Initial matrices and parameters
F = np.array([[1, dt], [0, 1]])  # State transition matrix
H = np.array([[1, 0]])           # Measurement matrix

# Kalman filter function
def kalman_filter(Q, R, measurements):
    x_hat, P = np.array([0, 0]), np.eye(2)  # Initial state and covariance
    estimated_positions, estimated_velocities = [], []
    residuals = []
    for z in measurements:
        # Prediction step
        x_hat = F @ x_hat
        P = F @ P @ F.T + Q

        # Measurement update step
        y = z - (H @ x_hat)
        S = H @ P @ H.T + R
        K = P @ H.T @ np.linalg.inv(S)
        x_hat = x_hat + K.flatten() * y  # Ensure K is 1D
        P = (np.eye(2) - K @ H) @ P

        estimated_positions.append(x_hat[0])
        estimated_velocities.append(x_hat[1])
        residuals.append(y)

    return np.array(estimated_positions), np.array(estimated_velocities), np.array(residuals)

# RMSE calculation function
def calculate_rmse(estimated, true):
    return np.sqrt(np.mean((estimated - true) ** 2))

# Method 1: Manual Tuning (Fixed)
Q_manual = np.array([[initial_process_noise_std_position**2, 0],
                     [0, initial_process_noise_std_velocity**2]])
R_manual = np.array([[measurement_noise_std**2]])

# Method 2: Empirical Estimation (Fixed)
def empirical_estimation_qr_batch(measurements, initial_Q, initial_R):
    est_positions, est_velocities, residuals = kalman_filter(initial_Q, initial_R, measurements)
    R_empirical = np.var(residuals)
    Q_empirical = initial_Q  # Keeping Q fixed for simplicity
    return Q_empirical, np.array([[R_empirical]])

# Method 3: Adaptive Filtering (Dynamic)
def kalman_filter_adaptive(measurements, initial_Q, initial_R, alpha=0.1):
    x_hat, P = np.array([0, 0]), np.eye(2)  # Initial state and covariance
    estimated_positions, estimated_velocities = [], []
    residuals = []
    R = initial_R.copy()
    
    for z in measurements:
        # Prediction step
        x_hat = F @ x_hat
        P = F @ P @ F.T + initial_Q

        # Measurement update step
        y = z - (H @ x_hat)
        S = H @ P @ H.T + R
        K = P @ H.T @ np.linalg.inv(S)
        x_hat = x_hat + K.flatten() * y  # Ensure K is 1D
        P = (np.eye(2) - K @ H) @ P

        estimated_positions.append(x_hat[0])
        estimated_velocities.append(x_hat[1])
        residuals.append(y)

        # Adaptive Update of R
        R = (1 - alpha) * R + alpha * np.array([[y**2]])

    return np.array(estimated_positions), np.array(estimated_velocities), np.array(residuals)

# Method 4: Expectation-Maximization (Iterative)
def expectation_maximization_qr(em_iterations, measurements, initial_Q, initial_R):
    Q_em, R_em = initial_Q.copy(), initial_R.copy()
    
    for _ in range(em_iterations):
        # E-step: Run Kalman Filter with current Q and R
        est_positions, est_velocities, residuals = kalman_filter(Q_em, R_em, measurements)
        
        # M-step: Update R based on residuals
        R_em = np.var(residuals)
        
        # Optionally, update Q here based on state estimates
        # For simplicity, we keep Q fixed
    return Q_em, np.array([[R_em]])

# Run Kalman filters with each method
results = {}

# 1. Manual Tuning
est_positions_manual, est_velocities_manual, _ = kalman_filter(Q_manual, R_manual, measurements)
position_rmse_manual = calculate_rmse(est_positions_manual, true_positions)
velocity_rmse_manual = calculate_rmse(est_velocities_manual, true_velocities)
results["Manual Tuning"] = {
    "estimated_positions": est_positions_manual,
    "estimated_velocities": est_velocities_manual,
    "position_rmse": position_rmse_manual,
    "velocity_rmse": velocity_rmse_manual
}
print(f"Manual Tuning - Position RMSE: {position_rmse_manual:.3f}, Velocity RMSE: {velocity_rmse_manual:.3f}")

# 2. Empirical Estimation
Q_empirical, R_empirical = empirical_estimation_qr_batch(measurements, Q_manual, R_manual)
est_positions_empirical, est_velocities_empirical, _ = kalman_filter(Q_empirical, R_empirical, measurements)
position_rmse_empirical = calculate_rmse(est_positions_empirical, true_positions)
velocity_rmse_empirical = calculate_rmse(est_velocities_empirical, true_velocities)
results["Empirical Estimation"] = {
    "estimated_positions": est_positions_empirical,
    "estimated_velocities": est_velocities_empirical,
    "position_rmse": position_rmse_empirical,
    "velocity_rmse": velocity_rmse_empirical
}
print(f"Empirical Estimation - Position RMSE: {position_rmse_empirical:.3f}, Velocity RMSE: {velocity_rmse_empirical:.3f}")

# 3. Adaptive Filtering
est_positions_adaptive, est_velocities_adaptive, _ = kalman_filter_adaptive(measurements, Q_manual, R_manual)
position_rmse_adaptive = calculate_rmse(est_positions_adaptive, true_positions)
velocity_rmse_adaptive = calculate_rmse(est_velocities_adaptive, true_velocities)
results["Adaptive Filtering"] = {
    "estimated_positions": est_positions_adaptive,
    "estimated_velocities": est_velocities_adaptive,
    "position_rmse": position_rmse_adaptive,
    "velocity_rmse": velocity_rmse_adaptive
}
print(f"Adaptive Filtering - Position RMSE: {position_rmse_adaptive:.3f}, Velocity RMSE: {velocity_rmse_adaptive:.3f}")

# 4. EM Algorithm
em_iterations = 5  # Number of EM iterations
Q_em, R_em = expectation_maximization_qr(em_iterations, measurements, Q_manual, R_manual)
est_positions_em, est_velocities_em, _ = kalman_filter(Q_em, R_em, measurements)
position_rmse_em = calculate_rmse(est_positions_em, true_positions)
velocity_rmse_em = calculate_rmse(est_velocities_em, true_velocities)
results["EM Algorithm"] = {
    "estimated_positions": est_positions_em,
    "estimated_velocities": est_velocities_em,
    "position_rmse": position_rmse_em,
    "velocity_rmse": velocity_rmse_em
}
print(f"EM Algorithm - Position RMSE: {position_rmse_em:.3f}, Velocity RMSE: {velocity_rmse_em:.3f}")

# Visualization of results
time = np.arange(0, num_steps * dt, dt)

# Plotting True Position vs Estimated Positions for each method
plt.figure(figsize=(14, 8))
for method, data in results.items():
    plt.plot(time, data["estimated_positions"], label=f"{method} Estimates")
plt.plot(time, true_positions, 'k--', label="True Position")
plt.scatter(time, measurements, color="orange", s=10, alpha=0.5, label="Measurements")
plt.xlabel("Time")
plt.ylabel("Position")
plt.legend()
plt.title("Kalman Filter Position Estimates by Method")
plt.show()

# Plotting True Velocity vs Estimated Velocities for each method
plt.figure(figsize=(14, 8))
for method, data in results.items():
    plt.plot(time, data["estimated_velocities"], label=f"{method} Estimates")
plt.plot(time, true_velocities, 'k--', label="True Velocity")
plt.xlabel("Time")
plt.ylabel("Velocity")
plt.legend()
plt.title("Kalman Filter Velocity Estimates by Method")
plt.show()
```
