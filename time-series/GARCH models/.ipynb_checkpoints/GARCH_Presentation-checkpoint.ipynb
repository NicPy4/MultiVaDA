  <!-- Title and definition box (appears first) -->
  <h1 style="font-size: 36px; text-align: center; color: #0d47a1; margin-bottom: 20px;">
    <strong>ARCH Models</strong>
  </h1>
  <p style="font-size: 22px; color: #424242; line-height: 1.6;">
    <strong>AutoRegressive Conditional Heteroskedasticity</strong> models are designed to handle time series data where the variance of errors is conditional on past values, i.e. not constant
  </p>
  
    
    
<div class="fragment fade-in" style="width: 100%; margin-top: 20px; display: flex; justify-content: space-between; align-items: flex-start;">

  <!-- First Column Wrapper -->
  <div style="width: 45%; display: flex; flex-direction: column;">
    <ul style="font-size: 20px; color: #424242; list-style-type: none; padding-left: 0; line-height: 1.8;">
      <li style="margin-bottom: 22px;"><strong>üìä Used to analyze and forecast volatility</strong></li>
      <li style="margin-bottom: 22px;"><strong>üìâ Assumes weak stationarity</strong></li>
    </ul>
  </div>

  <!-- Second Column Wrapper -->
  <div style="width: 45%; display: flex; flex-direction: column;">
    <ul style="font-size: 20px; color: #424242; list-style-type: none; padding-left: 0; line-height: 1.8;">
      <li style="margin-bottom: 22px;"><strong>‚è≥ Depends on past observations (lags)</strong></li>
      <li style="margin-bottom: 22px;"><strong>‚è∞ Has a "short memory" of past variance</strong></li>
    </ul>
  </div>
</div>

  <!-- Assumptions paragraph -->
  <p class="fragment fade-in" style="font-size: 22px; color: #424242; line-height: 1.6; margin-bottom: 20px;">
   Assuming zero mean, we get an ARCH(1) as:
  </p>

  <!-- Equation block -->
  <div class="fragment fade-in" style="background-color: #e1f5fe; padding: 20px; border-radius: 10px; box-shadow: 0px 4px 8px rgba(0,0,0,0.1);">
    <div style="text-align: center; font-size: 22px; color: #424242;">
        $$
        a_t = \nu_t \sqrt{\overbrace{\omega + \alpha_1 a_{t-1}^2}^{\sigma}} \qquad
        $$
      $$
      \begin{align}
        a_t \sim \mathcal{N}(0,1), \quad & \text{where } a_t \text{ is the return at time } t \\
        \nu_t \sim \mathcal{N}(0,1), \quad & \text{where } \nu_t \text{ is the error/noise term at time } t \\
        \sigma_0 =0, \quad & \text{where } \sigma_t \text{ is the variance/volatility at time } t
      \end{align}
      $$
    </div>
  </div>
    
  <div class="fragment fade-in" style="width: 100%; padding: 15px; border-radius: 10px; margin-top: 20px;">
    <p style="font-size: 20px; color: #333; text-align: center;">
      <em>Other distributions for the noise term can be considered, but the Gaussian is the most common.</em>
    </p>
  </div>
</div>

  <!-- Title and intro text (appears first) -->
  <h1 style="font-size: 36px; text-align: center; color: #0d47a1; margin-bottom: 20px;">
    <strong>GARCH Model</strong>
  </h1>
  <p style="font-size: 22px; color: #2c3e50; line-height: 1.6;">
    Generalized ARCH models use <span style="font-weight: bold; color: #1e88e5;">past observations and variances</span> to model and forecast volatility, particularly effective for <em>volatility clustering</em>.
  </p>
   
  <!-- Assumptions paragraph -->
  <p class="fragment fade-in" style="font-size: 22px; color: #2c3e50; line-height: 1.6; margin-bottom: 20px;">
   Assuming zero mean, the GARCH(1,1) model can be written as:
  </p>

  <!-- Equation block -->
  <div class="fragment fade-in" style="background-color: #e3f2fd; padding: 20px; border-radius: 10px; box-shadow: 0px 4px 8px rgba(0,0,0,0.1);">
    <div style="text-align: center; font-size: 22px; color: #2c3e50;">
        $$
        a_t = \nu_t \sqrt{\overbrace{\omega + \alpha_1 a_{t-1}^2 + \beta_1 \sigma_{t-1}^2}^{\sigma}} \qquad
        $$
      $$
      \begin{align}
        a_t \sim \mathcal{N}(0,1), \quad & \text{where } a_t \text{ is the return at time } t \\
        \nu_t \sim \mathcal{N}(0,1), \quad & \text{where } \nu_t \text{ is the error/noise term at time } t \\
        \sigma_0 = 0, \quad & \text{where } \sigma_t \text{ is the variance/volatility at time } t
      \end{align}
      $$
    </div>
  </div>
    
  <div class="fragment fade-in" style="width: 100%; margin-top: 20px;">
    <ul style="font-size: 22px; color: #2c3e50; list-style-type: none; padding-left: 0; line-height: 1.8;">
        <li style="margin-bottom: 22px;"><em>GARCH(1,1) is one of the most commonly used models in finance</em></li>
    </ul>
  </div>
    
</div>

  <!-- Title -->
  <h1 style="font-size: 36px; text-align: center; color: #0d47a1; margin-bottom: 25px;">
    <strong>Predicting Returns</strong>
  </h1>

  <!-- Main Explanation (appears first) -->
  <p style="font-size: 22px; color: #424242; line-height: 1.6; margin-bottom: 20px;">
    You can also model/predict returns \( r_t \) in addition to volatility by including the mean. It can be modeled in several different ways:
  </p>

  <!-- ARMA modeling equations and variance (appears next in the blue box) -->
  <div class="fragment fade-in" style="width: 100%; background-color: #e3f2fd; padding: 20px; border-radius: 10px; box-shadow: 0px 4px 8px rgba(0,0,0,0.1); margin-top: 20px;">
    <p style="font-size: 22px; color: #0d47a1; text-align: left; margin-bottom: 20px;">
      For instance:
    </p>
    <div style="font-size: 22px; color: #424242; text-align: center;">
      $$
      \begin{align}
         \textbf{Constant mean: } &\quad r_t = c + \epsilon_t \\
         \textbf{AR(1): } &\quad r_t = c + \phi r_{t-1} + \epsilon_t \\   
         \textbf{MA(1): } &\quad r_t = c + \epsilon_t + \theta \epsilon_{t-1}\\
         \textbf{ARMA(1,1): } &\quad r_t = c + \phi r_{t-1} + \epsilon_t + \theta \epsilon_{t-1}
      \end{align}
      $$
    </div>
    <p style="font-size: 20px; color: #424242; text-align: left; margin-bottom: 15px;">
      Where the variance is given by:
    </p>
    <div style="font-size: 22px; color: #424242; text-align: center;">
      $$
      \begin{align}
         \epsilon_t = \sigma_t \nu_t, \quad \sigma_t^2 = \omega + \alpha_1 \epsilon_{t-1}^2 + \beta_1 \sigma_{t-1}^2, \quad \nu_t \sim \mathcal{N}(0,1)
      \end{align}
      $$ 
    </div>
  </div>
</div>

  <h1 style="font-size: 36px; text-align: center; color: #1565c0; margin-bottom: 25px;">

  <div style="font-size: 22px; background-color: #e3f2fd; color: #42242;; line-height: 1.8; border-radius: 20px; box-shadow: 0px 4px 8px rgba(0,0,0,0.1);  padding: 20px; ">
    <p style="margin-bottom: 20px; line-height:1.2em">
      <strong style="color: #0d47a1;">Prepare the Data:</strong> 
        <br>
      Start by calculating the differences between successive values to capture the relative (percentile) changes over time
    </p>
    <p style="margin-bottom: 20px; line-height:1.2em">
      <strong style="color: #0d47a1;">Fit the Mean:</strong> 
      <br>
      Evaluate how the mean of the time series should be modeled, fitting an ARMA model if necessary
    </p>
    <p style="margin-bottom: 20px; line-height:1.2em">
      <strong style="color: #0d47a1;">Fit the GARCH Model:</strong> 
        <br>
      Use the residuals from the ARMA model, ot just the "raw" values to fit the GARCH model
    </p>
  </div>

```python
import numpy as np
import pandas as pd
from arch import arch_model
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# Load and preprocess data
data = pd.read_csv('data.csv')
close = data['Close'].values
log_returns = np.diff(np.log(close)) * 100  # Log returns scaled for better convergence
log_returns = np.diff(np.log(close)) * 100 
log_returns = pd.Series(log_returns)
data.head(-1)
```

```python
# Create a figure and two subplots next to each other
plt.figure(figsize=(16, 5))

# Plot original data (closing prices) on the first subplot
plt.subplot(1, 2, 1)
plt.plot(data['Date'], close, label='Closing Price')
plt.title('Original Closing Prices')
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(integer=True))  # Set x-axis for days
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.ylabel('Price')
plt.legend();

# Plot log returns on the second subplot
plt.subplot(1, 2, 2)
plt.plot(data['Date'][1:], log_returns , label='Log Returns', color='orange')
plt.title('Log Returns of Closing Prices')
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(integer=True))  # Set x-axis for days
plt.xlabel('Date')
plt.ylabel('Log Returns (%)')
plt.xticks(rotation=45)
plt.legend();
```

<h2 style="font-size: 36px; text-align: center; color: #1565c0; margin-bottom: 0px;">
    <strong>View ACF and PACF</strong>
  </h2>

```python
# Create a figure with two subplots to plot ACF and PACF side by side
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

sm.graphics.tsa.plot_acf(log_returns.values.squeeze(), lags=40, ax=ax[0])
ax[0].set_title('ACF of Log Returns')
sm.graphics.tsa.plot_pacf(log_returns.values.squeeze(), lags=40, ax=ax[1])
ax[1].set_title('PACF of Log Returns')

plt.tight_layout()
plt.show()
```

<h2 style="font-size: 36px; text-align: center; color: #1565c0; margin-bottom: 0px; ppadding-right:100px; padding-left: 100px; margin-left:100px; margin-right:100px">
    <strong>Lets fit a constant mean GARCH(1,1) model</strong>
  </h2>

```python
# Fit GARCH(1,1)
garch_model = arch_model(log_returns, mean='constant', vol='Garch', p=1, q=1) 
garch_fit = garch_model.fit (disp='off') # disp = 'off' means no printing during training
print(garch_fit.summary())

real_returns = log_returns[1:]  # Shifting because we lose one value in differencing
```

```python
# DONT USE IN PRESENTATION

# Plot the conditional volatility (Estimated vs Real)
plt.figure(figsize=(10, 5))
plt.plot(garch_fit.conditional_volatility, label='Estimated Volatility (GARCH(1,1))', color='orange')
plt.title('Estimated Conditional Volatility (GARCH(1,1))')
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(integer=True))  # Set x-axis for days
plt.xlabel('Time')
plt.ylabel('Volatility')
plt.legend()
plt.show()
```

<h2 style="font-size: 36px; text-align: center; color: #1565c0; margin-bottom: 0px;">
    <strong>Should we have modeled the mean?</strong>
  </h2>

```python
# Constant mean
garch_model_const = arch_model(log_returns, mean='constant', vol='Garch', p=1, q=1)
garch_model_const = garch_model_const.fit(disp='off') # disp = 'off' means no printing during training
print(garch_model_const.summary())

# Fit AR(1)-GARCH(1,1)
arima_model_ARMA11 = ARIMA(log_returns, order=(1, 1, 0))
arima_model_ARMA11 = arima_model_ARMA11.fit()
garch_model_ARMA11 = arch_model(arima_model_ARMA11.resid, mean='zero', vol='Garch', p=1, q=1)
garch_model_ARMA11 = garch_model_ARMA11.fit(disp='off') # disp = 'off' means no printing during training
print(garch_model_ARMA11.summary())

# Fit AR(1)-GARCH(1,1)
arima_model_AR1 = ARIMA(log_returns, order=(1, 0, 0))
arima_fit_AR1 = arima_model_AR1.fit()
garch_model_AR1 = arch_model(arima_fit_AR1.resid, mean='AR', vol='Garch', p=1, q=1, )
garch_model_AR1 = garch_model_AR1.fit(disp='off') # disp = 'off' means no printing during training
print(garch_model_AR1.summary())

# Fit MA(1)-GARCH(1,1)
arima_model_MA1 = ARIMA(log_returns, order=(0, 1, 0))
arima_model_MA1 = arima_model_MA1.fit()
garch_model_MA1 = arch_model(arima_model_MA1.resid, mean='AR', vol='Garch', p=1, q=1)
garch_model_MA1 = garch_model_MA1.fit(disp='off') # disp = 'off' means no printing during training
print(garch_model_MA1.summary())

# Fit AR(1)-GARCH(1,1)
arima_model_ARMA11 = ARIMA(log_returns, order=(1, 1, 0))
arima_model_ARMA11 = arima_model_ARMA11.fit()
garch_model_ARMA11 = arch_model(arima_model_ARMA11.resid, mean='AR', vol='Garch', p=1, q=1)
garch_model_ARMA11 = garch_model_ARMA11.fit(disp='off') # disp = 'off' means no printing during training
print(garch_model_ARMA11.summary())
```

```python
# Constant mean GARCH(1,1) model
garch_model_const = arch_model(log_returns, mean='constant', vol='Garch', p=1, q=1)
garch_model_const_fit = garch_model_const.fit(disp='off')
#print(garch_model_const_fit.summary())

# Fit AR(1)-GARCH(1,1)
arima_model_AR1 = ARIMA(log_returns, order=(1, 0, 0))
arima_fit_AR1 = arima_model_AR1.fit()
garch_model_AR1 = arch_model(arima_fit_AR1.resid, mean='AR', vol='Garch', p=1, q=1)
garch_model_AR1_fit = garch_model_AR1.fit(disp='off')
print(garch_model_AR1_fit.summary())

print("\nModel Comparison:")
print("Constant Mean - GARCH(1,1):")
print(f"AIC: {garch_model_const_fit.aic:.2f}, BIC: {garch_model_const_fit.bic:.2f}, Log-Likelihood: {garch_model_const_fit.loglikelihood:.2f}")

print("\nAR(1) - GARCH(1,1):")
print(f"AIC: {garch_model_AR1_fit.aic:.2f}, BIC: {garch_model_AR1_fit.bic:.2f}, Log-Likelihood: {garch_model_AR1_fit.loglikelihood:.2f}")
```

  <!-- Title -->
  <h2 style="font-size: 38px; text-align: center; color: #1565c0; margin-bottom: 25px;">
    <strong>Rolling Window Forecasting</strong>
  </h2>

  <!-- Explanation -->
  <p style="font-size: 22px; color: #424242; line-height: 1.6; margin-bottom: 20px;">
    Computes the <strong style="color: #1565c0;">true observed volatility</strong> in test data by calculating the standard deviation of returns over a specified time window.
  </p>

  <!-- Blue box for bullet points -->
  <div style="background-color: #e3f2fd; padding: 20px; border-radius: 10px; box-shadow: 0px 4px 8px rgba(0,0,0,0.1);">

<!-- Key Points -->
<ul style="font-size: 22px; color: #424242; list-style-type: none; padding-left: 0; line-height: 1.8;">
  <li class="fragment fade-in" style="margin-bottom: 15px;">
     <strong>Avoids lookback-bias:</strong> No use of future data! 
  </li>
  <li class="fragment fade-in" style="margin-bottom: 15px;">
    <strong>Less likely to overfit:</strong> Trains on different periods
  </li>
  <li class="fragment fade-in" style="margin-bottom: 15px;">
    <strong>Adaptive</strong>: Can catch changes
  </li>
  <li class="fragment fade-in" style="margin-bottom: 15px;">
    <strong>Two approaches: </strong> Expanding vs Fixed Window
  </li>
  <li class="fragment fade-in" style="margin-bottom: 15px;">
    <strong>Window size</strong> Case-specific and is bias-variance trade-off
  </li>
</ul>

  </div>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from arch import arch_model
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

# Split data into training and test sets
test_days = 50  # Number of days to use for the test set
train_size = len(log_returns) - test_days
train_data = log_returns[:train_size]
test_data = log_returns[train_size:]

dates = data['Date']
test_dates = dates[train_size:]

# Prepare arrays for rolling forecast
rolling_predictions = []
rolling_volatility = []

# Rolling forecasting on the test set
for i in range(len(test_data)):
    # Combine only training data and the observed test data up to this point, excluding the current test point
    combined_data = pd.concat([train_data, test_data[:i]], axis=0)

    # Step 1: Fit ARIMA on combined data (only on past data)
    arima_model = ARIMA(combined_data, order=(1, 0, 0))  # ARIMA(1,0,0) model
    arima_fit = arima_model.fit()

    # Step 2: Forecast 1-step ahead using ARIMA model (mean return prediction)
    predicted_return = arima_fit.forecast(steps=1)

    # Step 3: Fit GARCH(1,1) on residuals of ARIMA model for volatility
    residuals = arima_fit.resid
    garch_model = arch_model(residuals, vol='Garch', p=1, q=1)
    garch_fit = garch_model.fit(disp='off')

    # Step 4: Forecast 1-step ahead volatility using GARCH
    garch_forecast = garch_fit.forecast(horizon=1)
    predicted_vol = np.sqrt(garch_forecast.variance.iloc[-1, :].values)  # Get forecasted volatility (sigma)

    # Store predicted volatility
    rolling_volatility.append(predicted_vol[0])

    # Step 5: Simulate a random shock from normal distribution for the volatility-adjusted return
    shock = np.random.normal(0, 1)  # Standard normal distribution
    volatility_adjusted_return = predicted_return.iloc[0] + predicted_vol[0] * shock  # ARIMA return + GARCH shock

    # Store the volatility-adjusted return
    rolling_predictions.append(volatility_adjusted_return)

# Convert lists to arrays
rolling_volatility = np.array(rolling_volatility)
rolling_predictions = np.array(rolling_predictions)

# Create index for alignment (test_data index)
index_range = test_data.index
```

```python
def plot_volatility_and_returns(test_dates, rolling_volatility, test_data, rolling_predictions, rolling_window):
    true_volatility = test_data.rolling(window=rolling_window).std()
    
    # Create a figure with two subplots next to each other
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    test_dates = test_dates[1:]
    # Plot the predicted volatility vs true volatility in the first subplot
    axes[0].plot(test_dates, rolling_volatility, label='Predicted Volatility (GARCH)', color='orange', alpha=0.8)
    axes[0].plot(test_dates, true_volatility, label='True Volatility (Rolling Window)', color='blue', alpha=0.6)
    axes[0].set_title('Rolling Forecast of Volatility vs True Volatility (Test Set)')
    axes[0].set_xlabel('Date')
    axes[0].set_ylabel('Volatility')
    axes[0].legend()

    # Set x-axis to show only 8 evenly distributed dates
    tick_indices = np.linspace(0, len(test_dates) - 1, 8, dtype=int)
    tick_dates = test_dates.iloc[tick_indices]
    axes[0].set_xticks(tick_dates)
    axes[0].set_xticklabels(tick_dates, rotation=45)

    # Plot the actual returns vs predicted returns in the second subplot
    axes[1].plot(test_dates, test_data, label='Actual Returns', color='blue', alpha=0.6)
    axes[1].plot(test_dates, rolling_predictions, label='Predicted Returns (ARIMA + GARCH)', color='red', alpha=0.6)
    axes[1].set_title('Rolling Forecast of Returns: Predicted vs Actual (ARIMA + GARCH)')
    axes[1].set_xlabel('Date')
    axes[1].set_ylabel('Returns')
    axes[1].legend()

    # Set x-axis to show only 8 evenly distributed dates
    axes[1].set_xticks(tick_dates)
    axes[1].set_xticklabels(tick_dates, rotation=45)

    # Adjust layout to prevent overlap and display the plots
    plt.tight_layout()
    plt.show()
```

```python
def plot_volatility(test_dates, rolling_volatility, test_data, rolling_window):   
    # Calculate true volatility using a rolling window
    true_volatility = test_data.rolling(window=rolling_window).std()
    
    # Remove the first `rolling_window - 1` values from predicted volatility to align with true volatility
    aligned_volatility = rolling_volatility[rolling_window-1:]
    aligned_dates = test_dates[rolling_window:]
    true_volatility = true_volatility.dropna()  # Drop NaN values caused by rolling window

    fig, ax = plt.subplots(figsize=(10, 5))
    
    # Plot predicted volatility vs true volatility
    ax.plot(aligned_dates, aligned_volatility, label='Predicted Volatility (GARCH)', color='orange', alpha=0.8)
    ax.plot(aligned_dates, true_volatility, label='True Volatility (Rolling Window)', color='blue', alpha=0.6)
    
    ax.set_title('Rolling Forecast of Volatility vs True Volatility (Test Set)')
    ax.set_xlabel('Date')
    ax.set_ylabel('Volatility')
    ax.legend()

    # Set x-axis to show only 8 evenly distributed dates
    tick_indices = np.linspace(0, len(aligned_dates) - 1, 8, dtype=int)
    tick_dates = aligned_dates.iloc[tick_indices]
    ax.set_xticks(tick_dates)
    ax.set_xticklabels(tick_dates, rotation=45)

    # Adjust layout and display the plot
    plt.tight_layout()
    plt.show()
```
